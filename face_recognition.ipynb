### Set Up


Imports


```python
import sys
import cv2
import urllib
from google.colab.patches import cv2_imshow
```

Useful links 


```python
# Useful URLS
this_repo_url = 'https://github.com/madt2709/face-recognition/raw/main'
image_url = this_repo_url + '/images/abba.png'
casc_url = this_repo_url + '/images/haarcascade_frontalface_default.xml'

# Retrieving data
image_path = urllib.request.urlretrieve(image_url)[0]
cascade_path = urllib.request.urlretrieve(casc_url)[0]
```

### Face detection

```python
def face_detect(imgpath, nogui = False, cascasdepath = cascade_path):

    image = cv2.imread(imgpath)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier(cascasdepath)

    # Using common settings 
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor = 1.2,
        minNeighbors = 5,
        minSize = (30,30)
        )

    print("The number of faces found = ", len(faces))

    for (x,y,w,h) in faces:
        cv2.rectangle(image, (x,y), (x+h, y+h), (0, 255, 0), 2)

    if nogui:
        cv2.imwrite('test_face.png', image)
        return len(faces)
    else:
        # DisabledFunctionError: cv2.imshow() is disabled in Colab, because it causes Jupyter sessions
        # to crash; see https://github.com/jupyter/notebook/issues/3935.
        # As a substitution, consider using 
        # from google.colab.patches import cv2_imshow
        cv2_imshow(image)

if __name__ == "__main__":
    face_detect(image_path)
```
